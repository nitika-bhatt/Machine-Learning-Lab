```python
import numpy as np

X=np.array([0, 1, 2, 3, 3, 5, 5, 5, 6, 7, 7, 10])
Y=np.array([96, 85, 82, 74, 95, 68, 76, 84, 58, 65, 75, 50])

n=len(X)

sum_X=np.sum(X)
sum_Y=np.sum(Y)
sum_XY=np.sum(X*Y)
sum_X2 = np.sum(X**2)

m=((n*sum_XY)-(sum_X*sum_Y))/((n*sum_X)-sum_X**2)
print("The value of m: ", m)
C=(sum_Y-(m*sum_X))/n
print("The value of c is: ", C)

X_pred=9
Y_pred= (m*X_pred) +C
print("The predicted marks are: ", Y_pred) 
```
