```python
from sklearn.linear_model import LogisticRegression
```

```python
from sklearn.model_selection import KFold, cross_val_score
```

```python
from sklearn.metrics import accuracy_score
```

```python
np.random.seed(42)

weights=np.random.randint(20, 60, 100). reshape(-1, 1)
labels = (weights > 40).astype(int).ravel()

model=LogisticRegression()

KF=KFold(n_splits=5, shuffle=True, random_state=42)

scores= cross_val_score(model, weights, labels, cv=KF, scoring='accuracy')

print("accuracy_score", scores)
print("Mean accuracy: ", scores.mean())
print("Standard Deviation: ", scores.std())
```

```python
import matplotlib.pyplot as plt
model.fit(weights, labels)
x_range=np.linspace(weights.min(), weights.max(), 300).reshape(-1,1)
y_prob=model.predict_proba(x_range)[:, 1]

plt.scatter(weights, labels, c=labels, cmap="bwr", edgecolors="k", label="Data points")
plt.plot(x_range, y_prob, color="black", linewidth=2, label="Logistic Regression Curve")
plt.xlabel("Weight (kg)")
plt.ylabel("Probability of Obesity")
plt.title("Logistic Regression: Weight vs Obesity")
plt.legend()
plt.show()
```
